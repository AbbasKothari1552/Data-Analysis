# Read and convert into dictionary data


### Step 1: Read data 

```python
with open('Datasets/Crop_recommendation.txt', mode = 'r') as file:
    crop_data = file.read()
```

```python
crop_data
```

### Step 2 : Convert into string

```python
crop_data = crop_data.split()
```

```python
crop_data
```

### Step 3 : Now convert the individual string into a list

```python
crop_data = [row.split(',')  for row in crop_data]
```

```python
print(crop_data[:5])
```

### Step 4 : Seperate heading from the list 

```python
heading = crop_data[0]
print(heading)
```

```python
crop_data = crop_data[1:]
```

```python
print(crop_data[0])
```

```python
len(crop_data)
```

### Step 5 : Now match the heading with the lists values and store in the form of list

```python
crop_data_dict_list = []
for row in crop_data:
    dict = {}
    for key, value in zip(heading, row):
        dict[key] = value
    crop_data_dict_list.append(dict)
    
# Alternative and optimized code 
# code data = [dict(zip(heading,row)) for row in crop_data]
```
